```python
import pandas as pd
import matplotlib.pyplot as plt

# 3. Baca dataset
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

# 4. Hitung jumlah penonton per film
viewers = ratings.groupby('movieId').size().reset_index(name='jumlah_penonton')

# 5. Hitung rata-rata rating per film
avg_rating = ratings.groupby('movieId')['rating'].mean().reset_index(name='rata_rating')

# 6. Gabungkan data viewers + rating
film_data = pd.merge(viewers, avg_rating, on='movieId')

# 7. Gabung dengan data judul film
film_data = pd.merge(film_data, movies, on='movieId')

# 8. Urutkan berdasarkan jumlah penonton dan rating
top_films = film_data.sort_values(by=['jumlah_penonton', 'rata_rating'], ascending=False)

# 9. Tampilkan 10 film teratas
print(top_films[['title', 'jumlah_penonton', 'rata_rating']].head(10))

# 10. Visualisasi 10 film teratas (pakai matplotlib)
top10 = top_films.head(10)
plt.figure(figsize=(10,6))
plt.barh(top10['title'], top10['rata_rating'], color='skyblue')  # Gunakan matplotlib untuk bar chart horizontal
plt.title('10 Film Terbaik Berdasarkan Rating & Jumlah Penonton')
plt.xlabel('Rata-rata Rating')
plt.ylabel('Judul Film')
plt.tight_layout()
plt.show()
```
